Ты - эксперт в области бизнес аналитики. Твоя цель - максимально точно и подробно описать каждый кластер людей на основе доли их покупок. Попытайся вычленить максимально точное и глубокое описание их самых разнообразных возможных социально-демографических характеристик на основе топ 5 кластеров по доли покупок в каждой категории. На выходе жду текстовое описание по каждому номеру кластера.

```python
import pandas as pd
import numpy as np

df = pd.DataFrame(cross).set_index('socdem_cluster')

# Считаем доли покупок для каждого кластера
df_shares = df.div(df.sum(axis=1), axis=0)

# Функция для получения топ-5 и анти-топ-5 кластеров по доле покупок
def get_cluster_performance(df_shares, top_n=5):
    # Словари для хранения результатов
    top_categories_by_cluster = {}
    bottom_categories_by_cluster = {}
    
    # Проходим по всем категориям
    for category in df_shares.columns:
        # Топ-5 кластеров (наибольшие доли)
        top_clusters = df_shares[category].sort_values(ascending=False).head(top_n)
        # Анти-топ-5 кластеров (наименьшие доли)
        bottom_clusters = df_shares[category].sort_values(ascending=True).head(top_n)
        
        # Для каждого кластера из топ-5 добавляем категорию в его список
        for cluster in top_clusters.index:
            if cluster not in top_categories_by_cluster:
                top_categories_by_cluster[cluster] = []
            top_categories_by_cluster[cluster].append(category)
        
        # Для каждого кластера из анти-топ-5 добавляем категорию в его список
        for cluster in bottom_clusters.index:
            if cluster not in bottom_categories_by_cluster:
                bottom_categories_by_cluster[cluster] = []
            bottom_categories_by_cluster[cluster].append(category)
    
    return top_categories_by_cluster, bottom_categories_by_cluster

# Получаем результаты
top_categories, bottom_categories = get_cluster_performance(df_shares)

# Красивый вывод результатов по кластерам
print("ВЫВОД ПО КЛАСТЕРАМ: КАТЕГОРИИ ГДЕ КЛАСТЕР В ТОП-5 И АНТИ-ТОП-5")
print("="*80)

# Получаем все уникальные кластеры
all_clusters = sorted(set(list(top_categories.keys()) + list(bottom_categories.keys())))

for cluster in all_clusters:
    print(f"\nКЛАСТЕР {cluster}:")
    print("-" * 40)
    
    # Топ-5 категории для этого кластера
    top_cats = top_categories.get(cluster, [])
    if top_cats:
        print(f"  Входит в ТОП-5 в категориях:")
        for cat in sorted(top_cats):
            share = df_shares.loc[cluster, cat] * 100
            print(f"    • {cat}: {share:.5f}%")
    else:
        print(f"  Не входит в ТОП-5 ни в одной категории")
    
    # Анти-топ-5 категории для этого кластера
    bottom_cats = bottom_categories.get(cluster, [])
    if bottom_cats:
        print(f"  Входит в АНТИ-ТОП-5 в категориях:")
        for cat in sorted(bottom_cats):
            share = df_shares.loc[cluster, cat] * 100
            print(f"    • {cat}: {share:.5f}%")
    else:
        print(f"  Не входит в АНТИ-ТОП-5 ни в одной категории")
    
    # Статистика по кластеру
    avg_share = df_shares.loc[cluster].mean() * 100
    print(f"  Средняя доля по всем категориям: {avg_share:.5f}%")

# Дополнительная сводка
print("\n" + "="*80)
print("СВОДНАЯ СТАТИСТИКА:")
print(f"Всего кластеров: {len(all_clusters)}")
print(f"Кластеры с топ-5 категориями: {len(top_categories)}")
print(f"Кластеры с анти-топ-5 категориями: {len(bottom_categories)}")
```
